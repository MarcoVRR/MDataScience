## ¿Qué es GitHub?


Es un portal creado para alojar el código de las aplicaciones de cualquier desarrollador. La plataforma está creada para que los desarrolladores suban el código de aplicaciones y herramientas y que los uusuarios no puedan únicamente descargar la aplicación sino también entrar a su perfil para ler sobre ella o colaborar con su desarrollo. Fue lanzado en 2008 y se ha convertido en uno de los lugares más populares para almacenar y colaborar en proyectos de software.

## Características de GitHub



*   Repositorio: los proyectos de almacenan en repositorios que son espacios designados para organiza y gestionar el código fuente, la documentación y otros archivos relacionados.
*   Control de versiones: gitHub utuliza Git, un sistema de control de versiones distribuido. Permite realizar el seguimiento de cambios en el código fuente a lo largo del tiempo.
*   Colaboración: proporciona herramientas para que quienes desarrollan el proyecto trabajen junsot en él.
*   Integración continua: los desarrolladores pueden configurar servicios de integración continua para automatizar pruebas y construcciones cada vez que se realiza un cambio en el repositorio.

Las principales características de la plataforma es que ofrece las mejores características de ese tipo de servicios sin perder la simplicidad, y es una de las más utilizadas en el mundo por desarrolladores. Es multiplataforma, y tiene una gran cantidad de interfacesw de usuario.

En resumen, GitHub es un portal para gestionar las aplicaciones que utilizan el sistema Git. ## Otras características de GitHub

GitHub permite que los desarrolladores alojen proyectos creando repositorios de forma gratuita. Para usar esta opción los proyectos deberán ser de código abierto. Si no se desea que la aplicación sea de códido abierto, la herramienta también tiene una versión de pago para alojar proyectos de forma privada.

También se puede entrar a los proyectos de los demás y colaborar para mejorarlos. Un usuario puede opinar, comentar código y también mejorar el código. También pueden reportarse bugs o errores para que sean corregidos.

Entre otra de las herramientas que ofrece GitHub se puede crear una Wiki para cada proyecto, de esta forma puede ofrecerte información sobre él y las diferentes versiones.

Ofrece también una herramienta de seguimiwento de problemas y una de revisión de código, de esta forma se pueden crear debtates sonbre anotaciones para mejorar y optimizar el código.
